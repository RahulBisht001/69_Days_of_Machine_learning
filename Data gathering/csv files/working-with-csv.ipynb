### Sep Parameter

Generally in csv format the sep params is comma(,) and it is a default value. so we just need to override the default sep parameter like this -|

Some time it also happen that we don't have column heading names In that case the first row becomes the column heading. to overcome this ,we have a names attributes,its just a list for your column Headings.

```python
dataFrameOfTSV = pd.read_csv('/kaggle/input/csv-data/movie_titles_metadata.tsv',
                             sep ='\t',
                             names=['s.no','movie_name','release_year',
                                    'IMDB_rating','voting','generas'])
dataFrameOfTSV.head()

# We just need to add the additional param of (sep= \t) because of tsv file.
```

### 13. Handling Dates

By Default the Dates in Pandas are parsed as Strings so to put them into a date time format
we need to pass certain parameters.

* Note :-

*If you use the parse_dates parameter in the pd.read_csv() function to specify a column that is not actually a date, such as the name of a city column, it will not raise an error, but the resulting DataFrame will not have the desired datetime parsing for that column. Instead, Pandas will attempt to parse the values in the specified column as dates, and any non-date values will be treated as missing values (NaN) in the resulting DataFrame.*

> * *Read Docs for More Deep Understaing with Handling Dates.*

```python
# Uncomment the below line and see the type of date column. its and Object but with 
# parse_date we are converting all the column in a date time format which we pass as list.

# pd.read_csv('/kaggle/input/csv-data/IPL Matches 2008-2020.csv').info()

pd.read_csv('/kaggle/input/csv-data/IPL Matches 2008-2020.csv',
            parse_dates=['date']).info()
```

### 14. Converters

```python
def rename(name):
    if name == 'Royal Challengers Bangalore':
        return 'RCB'
    else :
        return name
```

```python
rename('Royal Challengers Bangalore')
rename('Mumbai Indians')
```

```python
pd.read_csv('/kaggle/input/csv-data/IPL Matches 2008-2020.csv',converters= {'team1':rename,'team2':rename})
```
