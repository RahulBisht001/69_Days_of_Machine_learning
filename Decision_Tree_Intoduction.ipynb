## **‚úÖDecision Trees**

A Decision tree is a flowchart-like structure in which each internal node represents a "test" on an attribute (e.g. whether a coin flip comes up heads or tails), each branch represents the outcome of the test, and each leaf node represents a class label (decision taken after computing all attributes).

How we can build our own Decision Tree.

**Pseduo code**

1.   Begin with the training dataset which should have some feature variables and **classification** or **Regression** output.

2.   Determine the **best feature** in the dataset to split the data on.

3. Split the data into subsets that contains the correct value for this **best feature**.This splitting basically defines a node on the tree i.e each node is a splitting point based on certain feature from our data.

4. Recusively generates new Tree node by using the subset of data created at step 3.

###ü§î**Entropy in Decision Tree**

In the context of Decision Trees, **entropy** is a measure of disorder or impurity in a node. Thus, a node with more variable composition, such as 2Pass and 2 Fail would be considered to have higher Entropy than a node which has only pass or only fail.

---
In layman terms **Entropy** is nothing but **the measure of disorder**

---
More Knowledge = less Entropy

### **ü§ØHow to Calulate Entropy**

*Mathematical Formula* :

$$E(S) = \sum_{i=1}^{c} -P(i) \log_{2}(P(i))$$
---

- where this \$P_{i}$  is nothing but frequentist probability of an element/class in the data.

* For Example if our data have only two class label yes or no.(eg: she love me or not [ Only yes and no 2 Categories ] )

$$E(D) = -p_{yes} \log_{2}(p_{yes})-p_{no} \log_{2}(p_{no})$$

- More the **Uncertainity** more the **Entropy**.
- For a 2 class problem the **min Entropty** is 0 and **max Entropy** is 1 but it not hold good more than 2 class. in more than 2 class problem the entropy can be more than 1.
- we can use both  $\log_{2}$  or $\log_{e}$  for calculating Entropies.

### **üìöInformation Gain**

**Information Gain** is a Matric used to train Decision Trees.Specifically ,this matric measures the **quality of split**.

---

The Information gain is based on the decrease in the entropy after the dataset is split on an attribute. Constructing a Decision Tree is all about finding attributes that return highest infomation Gain

$$ Information \ Gain = E(Parent) - Weighted Average * E(children)  $$

---
- Watch the CampusX Video for numerical

Steps for the Getting information gain


1.   Calculate Entropy of Parent $\ E_{parent}$
2.   Split the according to the attributes of the column which you have choose and the find out the
$\ E_{children}$
3.  Calculate Weighted Entropy of the children.

    $Weighted Entropy = (values \ in \ the \ child \ column / \ total \ values) * E_{children} $
    -
4. Calcute the **information gain** by formula.

5. Calculate the infomation gain for all the input columns and the algorithm will select the column with highest infomation gain.
6. Find information gain recursively

* Decision tree than applies a ***recursive greedy search algorithm*** in top to bottom fashion to find the information gain at each level of tree.
* Once the entropy comes 0 no more splitting is done.

### **üå≥Visualization of a Decision Tree**

```python
from sklearn.datasets import load_iris
import seaborn as sns
```

```python
# iris = sns.load_dataset('iris')
# iris.sample(5)

iris = load_iris()
X = iris.data
targetNames = iris.target_names
# print(targetNames)
y = iris.target
# y
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.2)
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
from sklearn.tree import DecisionTreeClassifier

clf = DecisionTreeClassifier()

clf.fit(X_train,y_train)
```

```python
y_pred = clf.predict(X_test)
# y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
from sklearn.tree import plot_tree
```

```python
from matplotlib.pylab import rcParams

rcParams['figure.figsize'] = 80,50

plot_tree(clf)
```
