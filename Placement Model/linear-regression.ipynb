## **Linear Regression**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/kaggle/input/placement/placement.csv')
# print(df)

df.head()

# df.head((10))

# .head() is a method used to display the first few rows of the DataFrame.
# It's a convenient way to quickly inspect the contents of the DataFrame 
# without printing the entire dataset, which can be useful when you're 
# working with large datasets.

# We can customize the number of rows displayed by passing an integer 
# argument to the .head() method. For instance, df.head(10) will display 
# the first ten rows of the DataFrame.

# and if the integer argument is greater than the total number of rows 
# in the DataFrame, it will simply display all available rows in the DataFrame.
```

```python
# matplotlib

plt.scatter(df['cgpa'],df['package'])
plt.xlabel('CGPA of the Student')
plt.ylabel('Package (in lpa)')
```

```python
X = df.iloc[:,0:1]
Y = df.iloc[:,-1]
# print(X)
# print(Y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train,Y_test = train_test_split(X,Y,test_size= 0.4,random_state = 2)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()

lr.fit(X_train,Y_train)
```

```python
X_test
```

```python
Y_test
```

```python
prediction = lr.predict(X_test.iloc[0].values.reshape(1,1))

print(prediction)
```

```python
# matplotlib
plt.scatter(df['cgpa'],df['package'])

plt.plot(X_train,lr.predict(X_train),color='red')
plt.xlabel('CGPA of the Student')
plt.ylabel('Package (in lpa)')
```

```python
m = lr.coef_ # this is the slope of the abobe Line
m
```

```python
intercept = lr.intercept_
intercept
```

### **Why it works and How it works ?**

Equation of a line
y = mx + b

here in Linear Regression -:
on the x axis we have CGPA and on the Y we have Package.

Now by the above "Best Fit Line Concept" we have m (slope) and b(intercept).
that's how our ML model is predicting the package for any new value of CGPA given to it.

e.g : CGPA = 8.58
hence y = mx + b

**package = 0.55193285 * 8.58 + (-0.8154867803356298)**
